```python
import pandas as pd
import matplotlib as plt
```

```python
data = pd.read_csv('state_level_summary.csv')
```

```python
data
```

```python
from pandas.io import sql
import MySQLdb as db

#Open connection to mysql database
con = db.connect('localhost','root','','FCC')

#Write dataframe to mysql database
data.to_sql(con=con, name='states', if_exists='replace', flavor='mysql')
```

```python
ax = data.percapita.plot(kind='bar',color='grey',grid=False)
labs = ax.get_xticklabels()
for lab in labs:
    lab.set_fontsize(6)
    lab.set_rotation(0)
ax.figure.set_figwidth(9)
ax.figure.set_figheight(3)
ax.figure.savefig('state_dist.png',dpi=200)
```
